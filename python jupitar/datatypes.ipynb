Numeric Data Types

```python
a = 5
b = 5.0
c = 2 + 4j

print(type(a))
print(type(b))
print(type(c))
```

Sequence Data Types

1. String

```python
s = 'Welcome to the Geeks World'
print(s)
print(type(s))

# access string with index
print(s[1])
print(s[-1])
```

2. List

```python
a = [1, 2, 3]
print(a)

b = ["Geeks", "For", "Geeks", 4, 5]
print(b[3])
print(b[-3])
```

3. Tuple

```python
t1 = (1,)
print(type(t1))

t2 = ('Geeks', 'For', 'Geeks', 1, 2)
print(t2[3])
print(t2[-3])
```

Boolean Data Type

```python
print(type(True))
print(type(False))
```

Truthy and Falsy Values

```python
if 1:
    print("1 is truthy")

if not 0:
    print("0 is falsy")
```

Dictionary Data Type

```python
d = {'a': 'Geeks', 'b': 'For', 'c': 'Geeks'}
print(d['a'])    
print(d.get('b'))
```
