User-defined Exceptions

```python
class InvalidMarksError(Exception):
    pass
def check_marks(marks):
    if marks<0 or marks>100:
        raise InvalidMarksError("Marks must be between 0 and 100")
    print("Valid Marks:", marks)
try:
    check_marks(120)
except InvalidMarksError as e:
    print(e)
```

Customizing Exception Classes

```python
class VotingAgeError(Exception):
    def _init_(self,age,msg="Age must be 18 or above"):
        self.age=age
        self.msg=msg
        super()._init_(self.msg)
    def _str_(self):
        return f"{self.age} -> {self.msg}"
def vote(age):
    if age<18:
        raise VotingAgeError(age)
    else:
        print("Eligible to vote")
try:
    vote(16)
except VotingAgeError as e:
    print(e)
```

Using Standard Exceptions as a Base Class

```python
class NetworkError(RuntimeError):
    def _init_(self,arg):
        self.args = (arg,)
try:
    raise NetworkError("Connection failed")
except NetworkError as e:
    print(e.args)
```

Real-World Example: Invalid Email Error

```python
class InvalidPhoneError(Exception):
    def _init_(self,phone,msg="Phone number must contain 10 digits"):
        self.phone=phone
        self.msg=msg
        super()._init_(self.msg)
    def _str_(self):
        return f"{self.phone} -> {self.msg}"
def set_phone(phone):
    if len(phone) != 0:
        raise InvalidPhoneError(phone)
    print(f"Phone number set to: {phone}")
try:
    set_phone("12345")
except InvalidPhoneError as e:
    print(e)
```
