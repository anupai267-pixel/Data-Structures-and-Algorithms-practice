For Loop

```python
n = 4
for i in range(0, n):
    print(i)
```

Iterating by Index of Sequences

```python
a = ["geeks", "for", "geeks"]
for idx in range(len(a)):
    print(a[idx])
```

While Loop

```python
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
```

Infinite While Loop

```python
while (True):
    print("Hello Geek")
```

Nested Loops

```python
for i in range(1, 5):
    for j in range(i):
        print(i, end=' ')
    print()
```
