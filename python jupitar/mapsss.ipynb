filter

```python
def odd(n):
    return n%2 !=0
a=[2,3,4,5,6]
s=filter(odd,a)
print(list(s))
```

```python
a=[1,2,3,4,5,6]
s=filter(lambda x:x % 2 ==0,a)
print(list(s))
```

Reduce

```python
from functools import reduce
a=[1,2,3,4,5]
r=reduce(lambda x,y:x *y, a)
print(r)
```

```python
from functools import reduce
b=[2,3,4,5,6]
s=reduce(lambda x,y: x if x>y else y,b)
print(s)
```

inner function

```python
def func1(msg):
    def func2():
        print(msg)
    func2()
func1("Python")
```

Decorators

```python
def decorator(func):
    def wrapper():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrapper

@decorator
def greet():
    print("Hello, World!")
greet()
```
