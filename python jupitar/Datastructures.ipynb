```python
a=[10,20,30]
print(a[2])
a.append(40)
print(a)
a.insert(2,50)
print(a)
a.extend([60,70])
print(a)
a[2]=90
print(a)
a.remove(20)
print(a)
a.pop()
print(a)
a.pop(3)
print(a)
del a[2]
print(a)
a.clear()
print(a)
```
