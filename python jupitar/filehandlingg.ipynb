File Handling

Creating a New File

```python
file = open("sample.txt", "w")
file.write("Hello World")
file.close()
```

Closing a File

```python
file=open("sample.txt","r")
file.close()
```

Reading a File

```python
file=open("sample.txt", "r")
content = file.read
print(content)
file.close()
```

Read one line

```python
file=open("sample.txt","r")
print(file.readline())
file.close()
```

Read all lines

```python
file=open("sample.txt","r")
print(file.readlines())
file.close()
```

Writing to a File

```python
file=open("sample.txt","w")
file.write("New Content")
file.close()
```

Appending to a File

```python
file=open("sample.txt","a")
file.write("\nPython File Handling")
file.close()
```

Using with Statement

```python
with open("sample.txt","r") as file:
    content=file.read()
    print(content)
```

Handling Exceptions When Closing a File

```python
try:
    file=open("sample.txt","r")
    content=file.read()
    print(content)
except FileNotFoundError as e:
    print("Error:",e)
finally:
    file.close()
```

Reading a file

Basic File Reading    :-                          Open the file        ,           Read content                &                     Close the file

```python
file=open("sample.txt","r")
content=file.read()
print(content)
file.close()
```

Reading a File Line by Line

```python
file = open("sample.txt", "r")
for line in file:
    print(line.strip())  
file.close()
```

```python
file = open("sample.txt", "r")
line = file.readline()
while line:
    print(line.strip())
    line = file.readline()
file.close()
```

Reading Binary Files

```python
with open("data.bin", "wb") as file:
    file.write(b"Hello Python")
```

Read the Binary File

```python
file=open("data.bin","rb")
content=file.read()
print(content)
file.close()
```

Reading Specific Parts of a File

```python
file = open("sample.txt", "r")
content = file.read(10)
print(content)
file.close()
```

Reading CSV Files

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Age"])
    writer.writerow(["Anu", 20])
    writer.writerow(["Raj", 21])

print("CSV file created.")
```

```python
import csv
import io
csv_data = """Year,Industry,Value
2014,Manufacturing,769400
2014,Manufacturing,48000
2014,Manufacturing,12"""
csvfile=io.StringIO(csv_data)
csvreader=csv.reader(csvfile)
for row in csvreader:
    print(row) 
```

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

Reading JSON Files

```python
import json
student={
    "name":"Anu",
    "age": 20
    }
with open("student.json","w") as file:
    json.dump(student, file)
print("JSON file created.")
```

```python
import json
with open("student.json","r") as jsonfile:
    data=json.load(jsonfile)
    print(data)
```
