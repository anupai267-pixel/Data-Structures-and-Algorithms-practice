1. Base Exceptiopn

```python
try:
    raise BaseException("This is a BaseException")
except BaseException as e:
    print(e)
```

2. Exception

```python
try:
    raise Exception("This is a generic exception")
except Exception as e:
    print(e)
```

3. Arithmetic Error

```python
try:
    raise ArithmeticError("Arithmetic error occured")
except ArithmeticError as e:
    print(e)
```

4. ZeroDivisionError

```python
try:
    result = 10/0
except ZeroDivisionError as e:
    print(e)
```

5. OverflowError

```python
import math
try:
    result = math.exp(1000)
except OverflowError as e:
    print(e)
```

6. FloatingPointError

```python
import numpy as np
np.seterr(all='raise')
try:
    np.divide(1,0)
except FloatingPointError as e:
    print("FloatingPointError caught:",e)
```

7. AssertionError

```python
try:
    assert 1==2, "Assert failed"
except AssertionError as e:
    print(e)
```

8. AttributeError

```python
class MyClass:
    pass
obj = MyClass()
try:
    obj.some_attribute
except AttributeError as e:
    print(e)
```

9. IndexError

```python
my_list =[1,2,3]
try:
    element = my_list[5]
except IndexError as e:
    print(e)
```

10. KeyError

```python
d={"key1": "value1"}
try:
    val=d["key2"]
except KeyError as e:
    print(e)
```

11. MemoryError

```python
try:
    li=[1]*(10*10)
except MemoryError as e:
    print(e)
```

12. NameError

```python
try:
    print(var)
except NameError as e:
    print(e)
```

13. OSError (and related errors)

```python
try:
    open("non_existent_file.txt")
except FileNotFoundError as e:
    print("FileNotFoundError caught:",e)
except OSError as e:
    print("OSError caught:",e)
```
