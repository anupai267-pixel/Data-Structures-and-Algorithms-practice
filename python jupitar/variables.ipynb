Assigning Values to Variables

```python
x = 5
name = "Alex"  
print(x)
print(name)
```

```python
a = b = c = 100
print(a, b, c)
```

```python
x, y, z = 1, 2.5, "Python"
print(x, y, z)
```

```python
x = 1
y = x
y = y + 1

print(x)
print(y)
```

```python
x = 10
del x
print(x)
```

Swapping Two Variables:

```python
a=10
b=5
a, b = b, a
print(a, b)
```
